```python
import numpy as np
import pandas as pd
#import tensorflow as tf
import matplotlib.pyplot as plt

from scipy.signal import savgol_filter
from IPython.display import display
```

```python
path_to_data = 'Данные СТМ Восточный Купол (от 29.01.17)\Куст 201\Скважина 2011.xlsx'
```

```python
# чтение данных и файла и возврат датафрейма
def get_df(path):
    file = pd.read_excel(path)
    file.rename(columns = {'Unnamed: 0' : 'Time Moment'}, inplace = True)
    return file
```

```python
# вывести график 4 переменных относительно времени
def draw_4_graphs(df):
    plt.figure(figsize=(20, 20))
    i = 1
    
    for col_name, data in df.items():
        if i != 1:
            plt.subplot(4, 1, i-1)
            
            #plt.plot(df['Time Moment'].values, data.values, '-')
            plt.plot(df['Time Moment'] , data, '-')
            plt.xticks(rotation = 45)
            plt.title(col_name)
        i+=1

    plt.show()
```

```python
# стандартизация данных
def standartize_data(df):
    i = True
    for col_name, data in df.items():
        if col_name != 'Time Moment': 
            df[col_name] = df[col_name] / df[col_name].max()
    return df
```

```python
# подготовка данных
def prepare_data(df):
    # удаление пропусков
    df.dropna(inplace = True)    
    print("After NAN delete: ", df.shape)
    
    # удаление нулей
    df = df.drop(labels = [0], axis = 0)
    print("After 0 delete: ", df.shape)
    
    # удаление выбросов
    Q1 =  df['TT2011-101 (Температура в ТП до УР)'].quantile(0.25)
    Q3 = df['TT2011-101 (Температура в ТП до УР)'].quantile(0.75)
    IQR = Q3 - Q1
    df.query("@Q1 - 1.5*@IQR < y < @Q3 + 1.5*@IQR").boxplot(column="y")
    
    df_IQR = df[(df['y'] > Q1) & (df['y'] < Q3)]
    df_IQR.boxplot(column="y")
    
    return df
```
