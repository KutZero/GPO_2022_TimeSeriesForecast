```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from scipy.signal import savgol_filter
from IPython.display import display
```

```python
path_to_data = 'Bush201\Well2011.xlsx'
```

```python
def get_df(path):
    file = pd.read_excel(path)
    #dframe = pd.DataFrame(file[target_name])
    return file
```

```python
firstDF = get_df(path_to_data)
display(firstDF)
```

```python
firstDF['PT2011-202 (Давление до УР)'].values
```

```python
firstDF = firstDF.drop(labels = [0], axis = 0)
display(firstDF)
```

```python
# Вывод графика изменения ошибки
plt.plot(firstDF['Unnamed: 0'].values, firstDF['PT2011-202 (Давление до УР)'].values)
plt.plot(firstDF['Unnamed: 0'].values, firstDF['PT2011-204 (Давление после УР)'].values)
#plt.plot(firstDF['Unnamed: 0'].values, firstDF['FT2011-301-1 (Расход в ТП до УР)'].values)
#plt.plot(firstDF['PT2011-202 (Давление до УР)'].values, firstDF['PT2011-204 (Давление после УР)'].values)
plt.title('Давление')
plt.ylabel('Давление после')
plt.xlabel('Давление до')
plt.legend(['Давление до', 'Давление после'], loc = 'upper left')
plt.show()
```

```python
y_filt_1 = savgol_filter(firstDF['PT2011-202 (Давление до УР)'].values, 61, 3)
y_filt_2 = savgol_filter(firstDF['PT2011-204 (Давление после УР)'].values, 61, 3)
```

```python
# Вывод графика изменения ошибки
plt.plot(firstDF['Unnamed: 0'].values, y_filt_1)
plt.plot(firstDF['Unnamed: 0'].values, y_filt_2)
#plt.plot(firstDF['Unnamed: 0'].values, firstDF['FT2011-301-1 (Расход в ТП до УР)'].values)
#plt.plot(firstDF['PT2011-202 (Давление до УР)'].values, firstDF['PT2011-204 (Давление после УР)'].values)
plt.title('Давление')
plt.ylabel('Давление после')
plt.xlabel('Давление до')
plt.legend(['Давление до', 'Давление после'], loc = 'upper left')
plt.show()
```
