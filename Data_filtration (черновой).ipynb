```python
# Вот так функция выглядит
# вообще она находится в "custom_modules\Data_visualization.py"
# вывести 4 графика автокорреляции (по графику на столбец)
# по возможности пользуйся функциями из библиотек
def draw_4_autocorrelation_graphs(df: pd.DataFrame):
    '''Рисует 4 графика автокорреляции (коррелограммы) для каждого значащего столбца датафрейма.

    :param df: arg1
    :type df: DataFrame
    '''

    plt.figure(figsize=(22, 20))
    i = 1
    columns = list(df.columns[1:])

    for col in columns:
        plt.subplot(4, 1, i)
        autocorrelation_plot(df[col]) # стандартная функция из pandas
        plt.xlabel('Time Moment')
        plt.ylabel(col)
        plt.title(col)
        i+=1

    plt.show()
```
