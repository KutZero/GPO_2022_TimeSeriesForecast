```python
def draw_4_frewuencis_graphs(df: pd.DataFrame):
    """Рисует 4 графика частоты встречающихся значений для каждого значащего столбца датафрейма.

    :param df: arg1
    :type df: DataFrame
    """
    columns = list(df.columns[1:])
    colors = ['b', 'm', 'g', 'y']
    
    fig, ax = plt.subplots(2,2)

    fig.set_figwidth(18)
    fig.set_figheight(14)
    
    colIter = iter(colors)
    
    ax = np.reshape(ax,(4,))
    
    itr = iter(ax)
    
    for col in columns:
        item = next(itr)
        
        n, bins, patches = item.hist(df[col], density=True, 
                                     bins=math.ceil((max(df[col]) - min(df[col])) / (0.2 / 10)),
                                     stacked=False, alpha=.7, color=next(colIter))
        
        item.text(bins[0] - 0.025, (max(n) + (max(n) / 10))/2, 'Левая граница', horizontalalignment='right', 
                  verticalalignment='center', color='r', fontsize=15, rotation='vertical')
        item.plot([bins[0], bins[0]], [0, max(n) + (max(n) / 10)], linestyle='-', color='r', linewidth=2)
        
        item.text(bins[-1] + 0.025, (max(n) + (max(n) / 10))/2, 'Правая граница', horizontalalignment='left', 
                  verticalalignment='center', color='r', fontsize=15, rotation='vertical')
        item.plot([bins[-1], bins[-1]], [0, max(n) + (max(n) / 10)], linestyle='-', color='r', linewidth=2)
        
        item.set_xlabel('Границы диапазонов (нормировано)', fontsize=15)
        item.set_ylabel('Частота встречи в выборке', fontsize=15)
        item.set_title(col[col.find('(') + 1:-1], fontsize=20)

        #  Устанавливаем интервал основных делений:
        item.xaxis.set_major_locator(ticker.MultipleLocator(0.2))
        item.yaxis.set_major_locator(ticker.MultipleLocator(max(n) / 10))

        #  Устанавливаем форматирование чисел основных делений:
        item.xaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))
        item.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

        #  Устанавливаем форматирование делений:
        item.xaxis.set_tick_params(which = 'major', labelsize = 15) #, labelrotation = 45)
        item.yaxis.set_tick_params(which = 'major', labelsize = 15)

        item.set_xlim(-0.1, 1.1);
        item.set_ylim(0, max(n) + (max(n) / 10));
        
        # Вывод и настройка сетки
        item.minorticks_on()
        item.grid(which='major', color = 'k', alpha = 0.6)
        item.grid(which='minor', color = 'gray', linestyle = ':')
        
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.2)
    plt.show()
```

```python
# вывести 4 графика зависимости значений столбца относительно времени (по графику на столбец)
def draw_4_graphs(df: pd.DataFrame):
    """Рисует 4 графика для каждого значащего столбца датафрейма.

    :param df: arg1
    :type df: DataFrame
    """
    
    columns = list(df.columns[1:])
    colors = ['b', 'm', 'g', 'r']
    
    fig, ax = plt.subplots(4,1)

    fig.set_figwidth(18)
    fig.set_figheight(24)
    
    colIter = iter(colors)
    itr = iter(ax)
    
    for col in columns:
        item = next(itr)
        step = (max(df[col]) - min(df[col]))/6
            
        item.plot([i for i in range(df[col].shape[0])], df[col], linestyle='-', color=next(colIter))

        item.set_xlabel('Номер измерения', fontsize=15)
        item.set_ylabel('Величина (нормировано)', fontsize=15)
        item.set_title(col[col.find('(') + 1:-1], fontsize=20)

        #  Устанавливаем интервал основных делений:
        item.xaxis.set_major_locator(ticker.MultipleLocator(400))
        item.yaxis.set_major_locator(ticker.MultipleLocator(step))

        #  Устанавливаем форматирование чисел основных делений:
        item.xaxis.set_major_formatter(ticker.FormatStrFormatter('%.d'))
        item.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

        #  Устанавливаем форматирование делений:
        item.xaxis.set_tick_params(which = 'major', labelsize = 15, labelrotation = 45)
        item.yaxis.set_tick_params(which = 'major', labelsize = 15)

        #item.set_ylim(-0.05, 1.05)
        item.set_ylim(min(df[col]) - step, max(df[col]) + step)
        
        # Вывод и настройка сетки
        item.minorticks_on()
        item.grid(which='major', color = 'k', alpha = 0.6)
        item.grid(which='minor', color = 'gray', linestyle = ':')
        
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
    plt.show()
```

```python
# вывсести 4 графика для двух датафреймов с наложением (по графику на столбец)
def draw_4_graphs_with_overlay_for_2_df(df1: pd.DataFrame, df2: pd.DataFrame):
    """
    Рисует 4 графика для двух датафреймов с наложением (по графику на столбец).
    Первый параметр - начальные данные, второй - обработанные данные.

    :param df1: arg1
    :type df1: DataFrame
    :param df2: arg2
    :type df2: DataFrame
    """

    columns = list(df1.columns[1:])
    
    fig, ax = plt.subplots(4,1)

    fig.set_figwidth(18)
    fig.set_figheight(24)

    itr = iter(ax)
    
    for col in columns:
        item = next(itr)
        item.plot([i for i in range(df1[col].shape[0])], df1[col], linestyle='-', 
                  color='r', label = "Начальные данные")
        item.plot([i for i in range(df2[col].shape[0])], df2[col], linestyle='-', 
                  color='b', label = "Обработанные данные")

        item.set_xlabel('Номер измерения', fontsize=15)
        item.set_ylabel('Величина (нормировано)', fontsize=15)
        item.set_title(col[col.find('(') + 1:-1], fontsize=20)

        #  Устанавливаем интервал основных делений:
        item.xaxis.set_major_locator(ticker.MultipleLocator(400))
        item.yaxis.set_major_locator(ticker.MultipleLocator(0.25))

        #  Устанавливаем форматирование чисел основных делений:
        item.xaxis.set_major_formatter(ticker.FormatStrFormatter('%.d'))
        item.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

        #  Устанавливаем форматирование делений:
        item.xaxis.set_tick_params(which = 'major', labelsize = 15, labelrotation = 45)
        item.yaxis.set_tick_params(which = 'major', labelsize = 15)

        # Вывод и настройка сетки
        item.minorticks_on()
        item.grid(which='major', color = 'k', alpha = 0.6)
        item.grid(which='minor', color = 'gray', linestyle = ':')
        item.legend(fontsize = 15, facecolor = "white")
        
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
    plt.show()
```
