```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display
from matplotlib import ticker
from pathlib import Path
```

```python
path_to_data = 'data\\Prepared data\\Run2\\X_data_hierarhical.xlsx' # куда сохранять X часть выборки
val_count = 8
```

```python
def get_df(path: str):
    df = pd.read_excel(path,index_col=[0,1])
    return df
```

```python
# Чтение данных из файла
first_df = get_df(path_to_data)
```

```python
# Что прочли из файла
display(first_df)
```

```python
first_df = first_df.loc[(slice(0,115),slice(0,64)), ~(first_df == '--').any()]
```

```python
first_df = first_df.drop(columns = ['detector_155','detector_182'],axis = 1)
```

```python
colormaps_list = list()

for i in range(val_count):
    colormaps_list.append(first_df.loc[(slice(0,115),[i]),:].to_numpy().astype(float))
```

```python
for i in range(val_count):
    print(colormaps_list[i].shape)
```

```python
fig, ax = plt.subplots(val_count,1)

fig.set_figwidth(18)
fig.set_figheight(36)

for item, colormap,i in zip(ax, colormaps_list, range(len(ax))):
    item.set_xlabel('Номер датчика', fontsize=15) 
    item.set_ylabel('Номер измерения', fontsize=15) 
    item.set_title(f'Значение времени № {i}', fontsize=15) 
    
    #  Устанавливаем интервал основных делений: 
    item.xaxis.set_major_locator(ticker.MultipleLocator(20)) 
    item.yaxis.set_major_locator(ticker.MultipleLocator(15)) 

    #  Устанавливаем форматирование чисел основных делений: 
    item.xaxis.set_major_formatter(ticker.FormatStrFormatter('%.d')) 
    item.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.d')) 

    #  Устанавливаем форматирование делений: 
    item.xaxis.set_tick_params(which = 'major', labelsize = 15, labelrotation = 45) 
    item.yaxis.set_tick_params(which = 'major', labelsize = 15) 

    item.pcolormesh(colormap)
    
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.8) 
plt.show()
```
