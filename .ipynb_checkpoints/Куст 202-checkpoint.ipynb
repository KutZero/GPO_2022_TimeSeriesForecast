```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('E:/4 семестр/ГПО/РПАС/Данные СТМ Восточный Купол (от 29.01.17)/Куст 202/Скважина 2021.xlsx')
df.head()
```

```python
#Графики по скважине 2021
df = pd.read_excel('E:/4 семестр/ГПО/РПАС/Данные СТМ Восточный Купол (от 29.01.17)/Куст 202/Скважина 2021.xlsx')
plt.rcParams['figure.figsize'] = (10,10)
plt.xticks(np.arange(0,df.index[-1],500))
plt.yticks(np.arange(-60,200,25))
plt.figure(1)
plt.title("Зависимость давления до УР от ТП")
plt.xlabel("ТП")
plt.ylabel("Давление до УР")
plt.plot(range(df.index[-1]-1), df['PT2021-202 (Давление до УР)'][1:df.index[-1]])
plt.legend(['PT2021-202 (Давление до УР)'])
plt.figure(2)
plt.plot(range(df.index[-1]-1), df['PT2021-204 (Давление после УР)'][1:df.index[-1]])
plt.legend(['PT2021-204 (Давление после УР)'])
plt.figure(3)
plt.plot(range(df.index[-1]-1), df['TT2021-101 (Температура в ТП до УР)'][1:df.index[-1]])
plt.legend(['TT2021-101 (Температура в ТП до УР)'])
plt.figure(4)
plt.plot(range(df.index[-1]-1), df['FT2021-301-1 (Расход в ТП до УР)'][1:df.index[-1]])
plt.legend(['FT2021-301-1 (Расход в ТП до УР)'])
plt.show()
```
